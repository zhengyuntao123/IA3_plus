## 处理MMLU数据集，构造提示词

```python
from datasets import load_dataset

mmlu = load_dataset("cais/mmlu", "all")
```

在lm-evaluation-harness\tests\test_prompt.py可以找到MMLU的提示词模板

```python
# 构建提示词模板
prompt_template="""The following are multiple choice questions (with answers) about {}.

{}
A. {}.
B. {}.
C. {}.
D. {}.
Answer:{}"""

EOS_TOKEN = tokenizer.eos_token

def formatting_prompt(examples):
    questions = examples['question']
    subjects = examples['subject']
    choices = examples['choices']
    answers = examples['answer']
    texts = []
    for question, subject, choice, answer in zip(questions, subjects, choices, answers):
        text = prompt_template.format(subject, question, choice[0], choice[1], choice[2], choice[3], answer) + EOS_TOKEN # 注意最后要加上EOS
        texts.append(text)
    return { "text" : texts, }
```

```python
training_data=mmlu['auxiliary_train'].map(formatting_prompt, batched= True)
```

```python
print(training_data['text'][0])
```

```python
# 计算所有text的长度
training_data = training_data.map(lambda x: {"text_len": len(x["text"])})
print(training_data[0]['text_len'])
print(len(training_data))
```

```python
plt.figure(figsize=(10, 3))
sns.histplot(training_data['text_len'], bins=50, kde=True)
plt.title('Distribution of Context Lengths')
plt.xlabel('Length of Context')
plt.ylabel('Frequency')
plt.show()
```

```python
# 观察到绝大多数样本的长度<4000，这里我们直接删除掉长度>4000的样本即可
training_data = training_data.filter(lambda x: x["text_len"] <= 4000)
len(training_data)
```

```python
plt.figure(figsize=(10, 3))
sns.histplot(training_data['text_len'], bins=50, kde=True)
plt.title('Distribution of Context Lengths')
plt.xlabel('Length of Context')
plt.ylabel('Frequency')
plt.show()
```

## IA3微调

```python
# 参考1：https://huggingface.co/docs/peft/conceptual_guides/ia3
# 参考2：https://github.com/huggingface/peft/blob/main/src/peft/tuners/ia3/config.py
from peft import IA3Config, get_peft_model,TaskType
```

```python
# 参考3: https://huggingface.co/docs/peft/main/package_reference/ia3
peft_config = IA3Config(
    task_type=TaskType.SEQ_CLS, target_modules=["k_proj", "v_proj", "down_proj"], feedforward_modules=["down_proj"]
)
```

```python
model = get_peft_model(model, peft_config)
model.print_trainable_parameters()
```

```python
from trl import SFTConfig, SFTTrainer
```

```python
training_args = SFTConfig(
    max_seq_length=512,
    learning_rate=3e-4,
    lr_scheduler_type="linear",
    per_device_train_batch_size=3,
    gradient_accumulation_steps=16,
    num_train_epochs=5,
    fp16=True,
    logging_steps=1,
    optim="adamw_8bit",
    weight_decay=0.01,
    warmup_steps=10,
    output_dir="output",
    seed=0,
)
```

```python
dataset = load_dataset("cais/mmlu", 'all')['auxiliary_train']
```

```python
trainer=SFTTrainer(
    model=model,
    tokenizer=tokenizer,
    train_dataset=training_data,
    dataset_num_proc=2,
    args=training_args
)
```

```python
trainer.train()
```

```python
model.save_pretrained("model/1B_finetuned_llama3.2_IA3")
tokenizer.save_pretrained("model/1B_finetuned_llama3.2_IA3")
```
